```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score,
    matthews_corrcoef,
    confusion_matrix,
    classification_report
)

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb

import joblib

from sklearn.preprocessing import StandardScaler
```

```python
import json
import os
```

#### Loading Dataset

```python
# from sklearn.datasets import load_breast_cancer

# data = load_breast_cancer(as_frame=True)
# df = data.frame

df = pd.read_csv("data/breast_cancer_wisconsin.csv")
df

```

```python
df = df.drop('Unnamed: 32', axis = 1)
```

```python
df
```

```python
print("\nData types of columns:\n")
df.dtypes
```

```python
print("\nStatistical summary:\n")
df.describe()
```

```python
print("Missing values per column:\n")
df.isna().sum()
```

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis'].map({'M':1, 'B':0})  # 1 = malignant, 0 = benign
print("Shape:", X.shape)
print("Target distribution:\n", y.value_counts())
```

#### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Combine features + target
test_df = X_test.copy()

# Convert 0/1 back to M/B used for clean code
test_df["diagnosis"] = y_test.map({1: "M", 0: "B"})

# Save CSV
test_df.to_csv("data/sample_test_data.csv", index=False)

print("Test dataset saved successfully.")
print("Shape:", test_df.shape)
```

#### Feature Scaling

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
joblib.dump(scaler, "scaler.pkl")
```

```python
# scaler = joblib.load("model/scaler.pkl")
# X_test_scaled = scaler.transform(X_test)
```

#### Initialising Models

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "XGBoost": xgb.XGBClassifier(eval_metric="logloss")
}
```

```python
results = []
model_reports = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    
    # Save model
    joblib.dump(model, f"{name.replace(' ', '_').lower()}.pkl")
    
    y_pred = model.predict(X_test_scaled)
    
    # Some models support predict_proba
    try:
        y_prob = model.predict_proba(X_test_scaled)[:,1]
    except:
        y_prob = y_pred  # fallback
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "AUC": roc_auc_score(y_test, y_prob),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
        "MCC": matthews_corrcoef(y_test, y_pred)
    })

    cm = confusion_matrix(y_test, y_pred)
    report = classification_report(
        y_test,
        y_pred,
        target_names=["Benign (0)", "Malignant (1)"],
        output_dict=True
    )

    model_reports[name] = {
        "confusion_matrix": cm.tolist(),  # JSON safe
        "classification_report": report
    }

os.makedirs("results", exist_ok=True)

with open("results/model_reports.json", "w") as f:
    json.dump(model_reports, f, indent=4)

results_df = pd.DataFrame(results)
print(results_df)
```

#### Report QC

```python
print("=== Confusion Matrix (Random Forest) ===")
rf = models["Random Forest"]
print(confusion_matrix(y_test, rf.predict(X_test_scaled)))

print("\n=== Classification Report (XGBoost) ===")
xg = models["XGBoost"]
print(classification_report(y_test, xg.predict(X_test_scaled)))
```
